 <b>2.11 Summary <b/>

As can be seen from the lineplot of oil rate production rates for each well all wells except for the F_4 (which did not produce any oil at all) and F_5 wells (which started out as a water injection well and later switched as production well) they all follow a typical decline curve. Another important factor to consider is that the direction of the correlation for a given variable for each well is not the same. An example of this is the pearson correlation for the average downhole pressure for each well. For some well there is a positive correlation with this variable but for some well it is negative. In the next notebook we will proceed to build models for all wells except the F_4 and F_5 to forecast their oil productions using different algorithms.

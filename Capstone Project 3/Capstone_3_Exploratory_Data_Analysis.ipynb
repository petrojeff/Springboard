```python
F_1=df.loc[df['NPD_WELL_BORE_NAME'] == '15/9-F-1 C' ]
F_11=df.loc[df['NPD_WELL_BORE_NAME'] == '15/9-F-11' ]
F_12=df.loc[df['NPD_WELL_BORE_NAME'] == '15/9-F-12' ]
F_14=df.loc[df['NPD_WELL_BORE_NAME'] == '15/9-F-14' ]
F_15=df.loc[df['NPD_WELL_BORE_NAME'] == '15/9-F-15 D' ]
F_4=df.loc[df['NPD_WELL_BORE_NAME'] == '15/9-F-4' ]
F_5=df.loc[df['NPD_WELL_BORE_NAME'] == '15/9-F-5' ]
```

```python
wells = [F_1,F_11,F_12,F_14,F_15,F_4,F_5]
```

 <b>2.4.1 Well oil production figures <b/>

```python
for well in wells:
    fig_dims = (10, 10)
    fig, ax = plt.subplots(figsize=fig_dims)
    g = sns.lineplot(x='new_date_column', y='BORE_OIL_VOL', data=well)
    g.set(xticks=[])
```
